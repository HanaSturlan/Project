```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets.samples_generator import make_blobs

#we create 40 separate points
x,y = make_blobs(n_samples=40, centers=2, random_state=20)
#40 data that have 2 centers - alligator or crocodile

#fit the model, don't regularize for illustration purposes
clf=svm.SVC(kernel='linear', C=1) # linear because it is 2-dimens.
clf.fit(x,y)

#display data i graph
plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap=plt.cm.Paired)
# x[:, 0] numpy - blobs return to numpy array (: all the rows, ,0 we are only gonna take first value)
```

```python
#predict new data

newData=[[3,4],[5,6]]
print(clf.predict(newData))
```

Splitting the dataset

```python
x=cancer.drop(['diagnosis'], axis=1)
```

```python
x.head()
```

```python
y=cancer['diagnosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=5)
```

```python
#training the model with SVM
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
svc_model=SVC()
```

```python
svc_model.fit(x_train,y_train)
```

```python
#evaluation
```

```python
y_predict=svc_model.predict(x_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
#pretty low acc
# can we  make it better by model improvisation?
```

```python
min_train=x_train.min()
```

```python
range_train=(x_train-min_train).max()
```

```python
x_train_scaled=(x_train-min_train)/range_train
```

```python
sns.scatterplot(x = x_train['area_mean'], y = x_train['smoothness_mean'], hue=y_train)
```

```python
sns.scatterplot(x = x_train_scaled['area_mean'], y = x_train_scaled['smoothness_mean'], hue=y_train)
```

```python
min_test=x_test.min()
range_test=(x_test-min_test).max()
x_test_scaled=(x_test-min_test)/range_test
```

```python
svc_model.fit(x_train_scaled,y_train)
```

```python
y_predict2=svc_model.predict(x_test_scaled)
y_predict2
```

```python
cm2=confusion_matrix(y_test,y_predict2)
```

```python
cm2
```

```python
#better accuracy!!!!
acc=(66+43)/(66+43+5)
acc
```

```python
print(classification_report(y_test,y_predict2))
```
