```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets.samples_generator import make_blobs

#we create 40 separate points
x,y = make_blobs(n_samples=40, centers=2, random_state=20)
#40 data that have 2 centers - alligator or crocodile

#fit the model, don't regularize for illustration purposes
clf=svm.SVC(kernel='linear', C=1) # linear because it is 2-dimens.
clf.fit(x,y)

#display data i graph
plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap=plt.cm.Paired)
# x[:, 0] numpy - blobs return to numpy array (: all the rows, ,0 we are only gonna take first value)
```

```python
#predict new data

newData=[[3,4],[5,6]]
print(clf.predict(newData))
```

```python
#first one iz on the left side, and antoher one is on the right
```

```python
#dig in deep!!
```

```python
#plot the decision function - we wanna know limits!!!

ax=plt.gca()
xlim=ax.get_xlim()
ylim=ax.get_ylim()

#creating grid
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy=np.vstack([XX.ravel(), YY.ravel()]).T # get them all in one array
Z = clf.decision_function(xy).reshape(XX.shape)
#clf - support vector

#plot decision boundary and margins

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1],
          linestyles=['--','-','--'])
#plot support vectors
ax.scatter(clf.support_vectors_[:, 0],
         clf.support_vectors_[:, 1], s=100,
          linewidth=1, facecolors='none')
plt.show()
```
